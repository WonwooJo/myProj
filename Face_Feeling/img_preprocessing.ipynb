```python
import cv2
import matplotlib.pyplot as plt
```

```python
gray = cv2.imread('images/eyes_L.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
gray
```

```python
plt.subplot(121),plt.imshow(gray)
plt.xticks([])
plt.yticks([])
plt.subplot(122),plt.hist(gray.ravel(), 255) 
plt.xticks([])
plt.yticks([])
plt.show()
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# only adaptiveThresholding
threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY, 11, 6)
plt.subplot(131), plt.imshow(gray)
plt.xticks([])
plt.yticks([])
plt.subplot(132), plt.imshow(threshold)
plt.xticks([])
plt.yticks([])
plt.subplot(133),plt.hist(threshold.ravel(), 255) 
plt.xticks([])
plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# otsu threshold
ret, threshold = cv2.threshold(gray,0, 255, cv2.THRESH_BINARY+
                            cv2.THRESH_OTSU)
plt.subplot(121), plt.imshow(gray)
plt.xticks([])
plt.yticks([])
plt.subplot(122), plt.imshow(threshold)
plt.xticks([])
plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# gausian blur + otsu threshold
blur = cv2.GaussianBlur(gray, (5,5),0)
ret, threshold = cv2.threshold(blur,0, 255, cv2.THRESH_BINARY+
                            cv2.THRESH_OTSU)
plt.subplot(131), plt.imshow(gray)
plt.xticks([])
plt.yticks([])
plt.subplot(132), plt.imshow(blur)
plt.xticks([])
plt.yticks([])
plt.subplot(133), plt.imshow(threshold)
plt.xticks([])
plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
